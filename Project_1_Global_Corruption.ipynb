## Variance

```python
display(df_CPI.loc['Fiji',:])   # Fiji is missing most CPI scores -- so we'll delete it later
```

```python
display(df_CPI)
```

```python
#  Calculate variance -- after managing NaN CPI score values

import numpy as np
import statistics as stat
import scipy.stats as sci


df_CPI['variance'] = 0         # Create 'variance' column

df_CPI = df_CPI.fillna(0)      # Temprorarily assign 0 to NaN values

for R in range(0,len(df_CPI)):

  for C in range(0,11):
    if df_CPI.iloc[R,C] == 0 :
       df_CPI.iloc[R,C] = stat.mean(df_CPI.iloc[R,0:11]) * 1.1

  df_CPI.iloc[R,11] = stat.variance(df_CPI.iloc[R,0:11])


df_CPI.head(20)

```

```python
# Sort by variance, copy highest values

df_CPI_sorted = df_CPI.sort_values(by=['variance'],ascending=False)

df_sorted = df_CPI_sorted.head(8)

df_sorted.drop(['Fiji'],inplace=True)    # too many missing values

df_CPI_sorted.head(8)

del df_CPI['variance']            # No longer needed
```

```python
# Plot countries with highest CPI variance

del df_sorted['variance']         # No longer needed

sortedT = df_sorted.transpose()

plt.figure(figsize=(18,10))

sns.lineplot(data=sortedT,linewidth=4,dashes=False,palette=['slateblue','orange','salmon','deepskyblue','seagreen','brown','fuchsia'],
             linestyle='solid').set(title='Corruption Perception Index - Greatest Variance')

plt.legend(ncols=4,fontsize='medium',bbox_to_anchor=(0.75,-0.06))

plt.show()
```
